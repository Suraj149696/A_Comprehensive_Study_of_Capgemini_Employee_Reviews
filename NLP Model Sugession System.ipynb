```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Load your Myntra fashion dataset into a Pandas DataFrame
# Assuming the product descriptions are in a column named 'description'

# Preprocess the text
product_descriptions = data2[0].apply(lambda x: x.lower())

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=10)  # Adjust the number of features as needed

# Fit and transform your product descriptions
tfidf_matrix = tfidf_vectorizer.fit_transform(product_descriptions)


# Get the feature names (keywords)
keywords = tfidf_vectorizer.get_feature_names_out()

# Sum the TF-IDF scores for each term across all product descriptions
tfidf_scores = tfidf_matrix.sum(axis=0)
print("TfIdf_Scores:", tfidf_scores)

# Sort the terms based on their TF-IDF scores
sorted_keywords = [keywords[i] for i in tfidf_scores.argsort()[0, ::-1]]

# Select the top N keywords
top_keywords = sorted_keywords[:20]
    
# Display or save the keywords
print("Top_Keywords:", top_keywords)
```

```python
import pandas as pd

# TF-IDF scores and top keywords
TfIdf_Scores = [[3.26048079, 1.82245488, 1.53675358, 0.89385032, 1.13352255, 1.03664078,
  1.07395755, 1.38012489, 1.53675358, 3.27680235]]
Top_Keywords = [['work', 'good', 'is', 'security', 'job', 'salary', 'like', 'not',
        'no', 'life']]

# Create a DataFrame
keyword_df = pd.DataFrame({'Description': data2[0],
                           'Keywords': Top_Keywords[0],
                           'TfIdf_Scores': TfIdf_Scores[0]})
keyword_df.index.name = 'Rank'

# Print the DataFrame
print(keyword_df)
```
